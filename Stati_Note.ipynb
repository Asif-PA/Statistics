# **Population And Sample**

Descriptive Statistics:

Focus: Descriptive statistics focus on describing and summarizing the characteristics of a sample or population.

Data Use: Descriptive statistics use the data directly from the sample or population being studied.

Purpose: The main goal is to provide clear and concise summaries of the data, including measures of central tendency (mean, median, mode) and measures of dispersion (range, variance, standard deviation).

Example: If you have a dataset containing the ages of all students in a school, descriptive statistics would calculate the average age of these students and the range of ages present in the dataset.

Inferential Statistics:

Focus: Inferential statistics focus on using information from a sample to make inferences or generalize conclusions about a population.

Data Use: Inferential statistics use sample data to draw conclusions or make predictions about a larger population.

Purpose: The main goal is to analyze the sample data to infer characteristics, parameters, or relationships within the population, while accounting for uncertainty.

Example: If you have a sample of 200 students from a school, inferential statistics might be used to estimate the average age of all students in the school district or to test hypotheses about the relationship between age and academic performance.

Key Differences:
Sample vs. Population: Descriptive statistics can describe both samples and populations directly. Inferential statistics, however, specifically use sample data to draw conclusions about populations.
Purpose: Descriptive statistics summarize data to provide insights into its features, while inferential statistics generalize findings beyond the sample to the larger population.
Statistical Techniques: Descriptive statistics typically include measures like averages and percentages, while inferential statistics involve techniques such as hypothesis testing and confidence intervals.
Practical Application:
In research and analysis, both descriptive and inferential statistics play crucial roles:

Descriptive Statistics: Used initially to understand and summarize the data collected from samples or populations.
Inferential Statistics: Applied subsequently to draw broader conclusions about populations based on sample data, while quantifying the uncertainty associated with these conclusions.
In summary, descriptive statistics focus on summarizing data from samples or populations, whereas inferential statistics use sample data to make broader generalizations or predictions about populations. Both are essential in statistical analysis for understanding and drawing conclusions from data.

When the variance of a data set is low, it indicates that the data points are close to the mean, whereas a high variance indicates that the observations are widely dispersed around the arithmetic mean and from one another.

While the central tendency, or average, tells you where most of your points lie, variability summarizes how far apart they are. This is important because the amount of variability determines how well you can generalize results from the sample to your population.

Low variability is ideal because it means that you can better predict information about the population based on sample data. High variability means that the values are less consistent, so its harder to make predictions.

Data sets can have the same central tendency but different levels of variability or vice versa. If you know only the central tendency or the variability, you cant say anything about the other aspect. Both of them together give you a complete picture of your data.

# **Distributions**

> **Normal Distrubution**

Normal Distribution, also known as Gaussian distribution, is a probability distribution that is symmetric about the mean, showing that data near the mean are more frequent in occurrence than data far from the mean. In graphical form, the normal distribution will appear as a bell curve.

**Key Characteristics:**

Symmetry: The distribution is symmetric about the mean.

Mean, Median, and Mode: All are equal and located at the center of the distribution.

Bell Curve: The shape of the distribution is bell-shaped.

68-95-99.7 Rule: About 68% of the data falls within one standard deviation of the mean, 95% within two, and 99.7% within three.

**Importance in Data Science**

Real-World Phenomena: Many real-world phenomena follow a normal distribution, such as heights, test scores, and measurement errors.

Statistical Methods: Many statistical tests and methods assume normality, making it essential to understand and use this distribution.

```python
import seaborn as sns

# Parameters for the normal distribution
mu, sigma = 100, 15  # mean and standard deviation

# Generate 1000 random samples from a normal distribution with mean=100 and std=15
data = np.random.normal(mu, sigma, 1000)

# Plotting
sns.histplot(data, kde=True, color='g')
plt.title('Normal Distribution')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

**python code example for converting a normal distrubutiopn to standard normal distrubution**

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate random data points from a normal distribution
mu = 70  # Mean
sigma = 10  # Standard deviation
data_points = np.random.normal(mu, sigma, 1000)

# Standardize the data points
standardized_data_points = (data_points - mu) / sigma

# Create the plot
plt.figure(figsize=(12, 6))

# Plot the original normal distribution
plt.subplot(1, 2, 1)
sns.histplot(data_points, kde=True)
plt.axvline(mu, color='r', linestyle='--', label='Mean')
plt.title('Original Normal Distribution')
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.legend()

# Plot the standard normal distribution
plt.subplot(1, 2, 2)
sns.histplot(standardized_data_points, kde=True)
plt.axvline(0, color='r', linestyle='--', label='Mean (0)')
plt.title('Standard Normal Distribution')
plt.xlabel('Z-scores')
plt.ylabel('Frequency')
plt.legend()

plt.tight_layout()
plt.show()
```

**Random Variable**

A random variable is a variable whose possible values are numerical outcomes of a random phenomenon. It is a function that assigns a numerical value to each outcome in the sample space of a random experiment.

**Types of Random Variables**

**Discrete Random Variable:** Takes on a countable number of distinct values. Examples include the number of heads in 10 coin flips, the number of students in a class, etc.

**Continuous Random Variable:** Takes on an infinite number of possible values. Examples include the height of students, the amount of time taken to run a race, etc.

**Probability Distribution Function**

A probability distribution function describes how probabilities are distributed over the values of the random variable.

**Types of Probability Distributions**

**Discrete Probability Distribution:** Deals with discrete random variables. Example: Binomial distribution, Poisson distribution.

**Continuous Probability Distribution:** Deals with continuous random variables. Example: Normal distribution, Exponential distribution.

**A probability distribution is a list of all of the possible outcomes of a random variable along with their corresponding probability values.**

In many scenarios, the number of outcomes can be much larger and hence a table would
be tedious to write down. Worse still, the number of possible outcomes could be infinite,
in which case, good luck writing a table for that.

**Famous Probability Distributions**

**Binomial Distribution:** Models the number of successes in a fixed number of independent Bernoulli trials.

**Poisson Distribution:** Models the number of events occurring within a fixed interval of time or space.

**Normal Distribution:** Describes a continuous random variable with a symmetric, bell-shaped probability distribution.

**Exponential Distribution:**Models the time between events in a Poisson process.
Uniform Distribution: All outcomes are equally likely.

**PMF stands for Probability Mass Function.**

 It is a mathematical function that describes the probability distribution of a discrete random variable.

The PMF of a discrete random variable assigns a probability to each possible value of the random variable. The probabilities assigned by the PMF must satisfy two conditions:

1. The probability assigned to each value must be non-negative (i.e., greater
than or equal to zero).

2. The sum of the probabilities assigned to all possible values must equal 1.
